```python
prediction_days = 10
```

```python
df[f"MA_{short}_days_{prediction_days}_days_prediction"] = np.NaN
```

```python
def make_df(prediction_days, start_date):
    return [ start_date+i for i in range(prediction_days) ]
```

```python
for idx in range(short, len(df), prediction_days):
    for i in range(prediction_days):
        if idx+i >= len(df):
            break
        close = make_df(short-i, idx+i-short)
        MA = make_df(i,idx)
        avg = pd.concat([df["Adj Close"].iloc[close],df[f"MA_{short}_days_{prediction_days}_days_prediction"]])
        df.iat[idx+i, df.columns.get_loc(f"MA_{short}_days_{prediction_days}_days_prediction")] = avg
```
